## VisualizingData
<strong>Sales Data Analysis</strong> <br>
Analyzing monthly product sales in 3 different ways. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import CSV of monthly sales
df = pd.read_csv("MonthlyProductSales.csv", encoding = "ISO-8859-1")
```

```python
df
```

### i. Default plot

```python
# plotting total sales by month

df.groupby(df['Month of Order Date']).sum().plot()
```

### ii. Histogram plot

```python
# plotting histogram over similar data

df.groupby(df['Month of Order Date']).sum().plot(kind='hist', bins=10)
```

### Horizontal Bar plot

```python
# groupby: Product name
# sort by: Sales sum
# fetching largest sales from bottom because of ascending order
# plot that data using horizontal bar visualization

df.groupby(df['Product Name']).sum().sort_values(by='Sales').tail(10).plot(kind='barh')
```
